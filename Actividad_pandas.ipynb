```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("alumnos.csv", encoding = 'latin-1')
```

```python
df.info()
print("Numero de datos: " , len(df))
```

## Rangos

```python
df.describe()
```

## Media

```python
for col in df.columns:
    try:
        print('Media ', col, ':' ,df[col].mean())
    except:
        print("", end = "")  
```

## Mediana

```python
for col in df.columns:
    try:
        print('Mediana ', col, ':' ,df[col].median())
    except:
        print("", end = "")  
```

## Desviación estandar

```python
for col in df.columns:
    try:
        print('std ', col, ':' ,df[col].std())
    except:
        print("", end = "")  
```

## Conclusiones de los datos

En promedio el numero de hombres y mujeres por grado es casi igual, tiene pequeñas variaciones de hasta 1 alumno por grado.

La mediana es el punto medio del conjunto de datos. El valor de este punto medio es el punto en el cual la mitad de las observaciones esta por encima del valor y la otra mitad esta debajo del valor, por lo que podemos obervar que la media es 4 en los alumnos de  primer grado, esto de 14000 registros, por lo que existen varios salones con menos de 4 alumnos hombres y mujeres.

Una desviación estandar baja indica que la mayor parte de los datos de una muestra tiende a estar agrupados cerca del valor de la media, mientras que una desviación estandar alta indica que los valores se extienden sobre un rango de valores mas amplio, en este caso podemos observar que la desviacion estandar llega a tener un valor de hasta 52 alumnos en el total de alumnos de 1°, por lo que consideramos eso como una desviación estandar, esto quiere decir que el numero de alumnos de cada salon llega a variar mucho entre cada escuela.
